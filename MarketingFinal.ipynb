```python
Marketing2 = Marketing[['Year_Birth','Dt_Customer','Education','Marital_Status','Kidhome','Teenhome','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds','NumDealsPurchases','NumWebVisitsMonth','NumWebPurchases','NumCatalogPurchases','NumStorePurchases']]
```

```python
# Creating a new column for age.
```

```python
Marketing2['Year_Birth'] = pd.to_datetime(Marketing2['Year_Birth'], format ='%Y')
Marketing2['Dt_Customer'] = pd.to_datetime(Marketing2['Dt_Customer'])
```

```python
Marketing2["Dt_Customer"].dt.year.min() , Marketing2["Dt_Customer"].dt.year.max()
```

```python
Marketing2['CustomerAge'] = (Marketing2["Dt_Customer"].dt.year.max()) - (Marketing2['Year_Birth'].dt.year)
```

```python
Marketing2.head()
```

```python
# Data Exploration
```

```python
Marketing2.Education.value_counts()
```

```python
Marketing2.Marital_Status.value_counts()
```

```python
# Clustering some columns
```

```python
Marketing2['Couple']= Marketing2['Marital_Status'].replace({'Married':'Yes', 'Together':'Yes', 'Absurd':'No', 'Widow':'No', 'YOLO':'No', 'Divorced':'No', 'Single':'No','Alone':'No'})
```

```python
Marketing2['LvlEducation']=Marketing2['Education'].replace({'Basic':'Basic','2n Cycle':'2n Cycle', 'Graduation':'Graduate', 'Master':'Postgraduate', 'PhD':'Postgraduate'})
```

```python
Marketing2["Children"] = Marketing2["Kidhome"] +  Marketing2["Teenhome"]
```

```python
Marketing2.head()
```

```python
Marketing3 = Marketing2[['CustomerAge','Couple','LvlEducation','Children','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']]
```

```python
Marketing3.head()
```

```python
Marketing3.CustomerAge.value_counts()
```

```python
Marketing3.CustomerAge.describe()
```

```python
# Creating age range
```

```python
Marketing3['AgeRange'] = pd.cut(x = Marketing3['CustomerAge'], bins = [18,25,45,60,121],
                         labels = ['18+','Adult','MiddleAge','Senior'])
```

```python
Marketing3.head()
```

```python
Marketing3['TotalMntSpent'] = Marketing3['MntWines'] + Marketing3['MntFruits'] + Marketing3['MntMeatProducts'] + Marketing3['MntFishProducts'] + Marketing3['MntSweetProducts'] + Marketing3['MntGoldProds']
```

```python
Marketing4=Marketing3[['AgeRange','Couple','LvlEducation','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds','TotalMntSpent']]
```

```python
Marketing4.head()
```

```python
Marketing5 = Marketing4.groupby(['AgeRange','LvlEducation','Couple'])
```

```python
Marketing5.first()
```

```python
# Correlation Matrix
```

```python
Marketing5.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('Blues'), axis=1)
```
